```python
#Making imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([1,10,20,40,50,70,80,90,120])
y=np.array([3,20,90,110,130,170,150,200,260])
plt.scatter(x,y)
plt.show()
```

```python
#Building the model
m=0
c=0
learning_rate=0.0001
numberofiterations=1000
n=float(len(x))
for i in range(len(x)):    
    y_pred=m*x+c
    D_m=(-2/n)*sum(x*(y-y_pred))
    D_c=(-2/n)*sum(y-y_pred)
    m=m-learning_rate*D_m
    c=c-learning_rate*D_c
print(m,c)
```

```python
#Making predictions
y_pred=m*x+c
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)],color='yellow')
plt.show()
```
